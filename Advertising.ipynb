```python
from sklearn.model_selection import train_test_split
```

```python
ad_topic = pd.get_dummies(ad_data['Ad Topic Line'],drop_first=True)
ad_topic
```

```python
ad_data.drop(['Ad Topic Line','City','Country','Timestamp'],axis=1,inplace=True)
```

```python
ad_data = pd.concat([ad_data,ad_topic],axis=1)
ad_data
```

```python
y = ad_data['Clicked on Ad']
```

```python
X = ad_data.drop(['Clicked on Ad'],axis=1)
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,predictions))
```

```python
pred = model.predict(X)
```
